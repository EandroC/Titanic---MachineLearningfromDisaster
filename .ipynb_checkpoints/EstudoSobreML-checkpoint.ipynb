```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv("Dados/train.csv", sep=",")
test = pd.read_csv("Dados/test.csv", sep=",")
```

```python
train.head()
```

```python
modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

```python
train['Sex'].value_counts()
```

```python
#Esse modelo apenas trabalha com dados numéricos então é necessário trocar valores da coluna Sex. obs: no train e test
def transformar_sexo(valor):
    if valor == 'female':
        return 1
    else:
        return 0
#map pega a função entre parenteses e aplica linha a linha na coluna selecionada.
train['Sex_binario'] = train['Sex'].map(transformar_sexo)
```

```python
train.head()
```

```python
variaveis = ['Sex_binario', 'Age']
```

```python
X = train[variaveis]
y = train['Survived']
```

```python
X.head()
```

```python
y.head()
```

```python
#normalmente modelos de MLs não vão funcionar com valores NaN ou valores muito grandes(ex float)
#fillna vai substituir todos os valores vazios por -1(no exemplo) obs: no train e test
#-1 pq tanto nas duas colunas usadas ele será interpretado como um valor fora do comum
X = X.fillna(-1)
```

```python
modelo.fit(X, y)
```

```python
test['Sex_binario'] = test['Sex'].map(transformar_sexo)
```

```python
X_prev = test[variaveis]
X_prev = X_prev.fillna(-1)
X_prev.head()
```

```python
p = modelo.predict(X_prev)
p
```

```python
sub = pd.Series(p, index=test['PassengerId'], name='Survived')
sub.shape
```

```python
#Fazendo o arquivo
sub.to_csv('Modelos/ModelosEstudo/modeloEstudo1.csv', header= True)
```

### Validação cruzada

```python
X_falso
```

```python
from sklearn.model_selection import KFold
```

```python
#?KFold
kf = KFold(3, shuffle=True, random_state=0)
for linhas_treino, linhas_valid in kf.split(X_falso):
    print('Treino: ', linhas_treino)
    print('Validação: ', linhas_valid)
    print()
    
```

```python
#?KFold
#random_state = seed, pode ficar trocando
resultados =[]
for rep in range(10):
    print('Rep: ', rep)
    kf = KFold(3, shuffle=True, random_state=0)

    for linhas_treino, linhas_valid in kf.split(X):
        print('Treino: ', linhas_treino.shape[0])
        print('Validação: ', linhas_valid.shape[0])
        print()

        X_treino, X_valid = X.iloc[linhas_treino], X.iloc[linhas_valid]
        y_treino, y_valid = y.iloc[linhas_treino], y.iloc[linhas_valid]

        modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
        modelo.fit(X_treino, y_treino)

        p = modelo.predict(X_valid)

        acc = np.mean(y_valid == p)
        resultados.append(acc)

        print('Acc: ', acc)
        print()
    
    #print(X_treino.head())
```

```python
np.mean(resultados)
```

```python
from sklearn.model_selection import RepeatedKFold
```

```python
resultados = []
kf = RepeatedKFold(n_splits=2, n_repeats=10, random_state=10) 

for linhas_treino, linhas_valid in kf.split(X):
    print( 'Treino: ', linhas_treino.shape[0])
    print('Validação: ', linhas_valid.shape[0])
    print()
    
    X_treino, X_valid = X.iloc[linhas_treino], X.iloc[linhas_valid]
    y_treino, y_valid = y.iloc[linhas_treino], y.iloc[linhas_valid]
    
    modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
    modelo.fit(X_treino, y_treino)
    
    p = modelo.predict(X_valid)
    
    acc = np.mean(y_valid == p)
    resultados.append(acc)
    
    print('Acc: ', acc)
    print()
    
    #print(X_treino.head())
```

### Novo modelo

```python
resultados = []
kf = RepeatedKFold(n_splits=2, n_repeats=10, random_state=10) 

for linhas_treino, linhas_valid in kf.split(X):
    print( 'Treino: ', linhas_treino.shape[0])
    print('Validação: ', linhas_valid.shape[0])
    print()
    
    X_treino, X_valid = X.iloc[linhas_treino], X.iloc[linhas_valid]
    y_treino, y_valid = y.iloc[linhas_treino], y.iloc[linhas_valid]
    
    modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
    modelo.fit(X_treino, y_treino)
    
    p = modelo.predict(X_valid)
    
    acc = np.mean(y_valid == p)
    resultados.append(acc)
    
    print('Acc: ', acc)
    print()
    
```

```python
# P previsão que o modelo fez pros pois vou checar o modelo anterior e ver possiveis mudanças
X_valid_check = train.iloc[linhas_valid].copy()
X_valid_check['p'] = p
X_valid_check.head()
```
